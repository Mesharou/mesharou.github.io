```python

def rankine(r,rm,vm):
    v = copy(r)
    v = vm* r/rm
    v[np.abs(r)>rm] = vm* (r[np.abs(r)>rm]/rm)**(-1)
    return v

def gaussian(r,rm,vm):
    v = copy(r)
    v = vm* r/rm * np.exp(-r**2/(2*rm**2)) / np.exp(-1/2)
    return v
```
