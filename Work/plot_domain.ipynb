```python
def plot_domain(simul,ax):
    domaincolor = 'DimGray'
    domainalpha = 1.
    ax.plot(simul.x[-1,:], simul.y[-1,:], color=domaincolor,alpha=domainalpha, linewidth=5, transform=ccrs.PlateCarree())
    ax.plot(simul.x[:,0], simul.y[:,0], color=domaincolor,alpha=domainalpha, linewidth=5, transform=ccrs.PlateCarree())
    ax.plot(simul.x[0,:], simul.y[0,:], color=domaincolor,alpha=domainalpha, linewidth=5, transform=ccrs.PlateCarree())
    ax.plot(simul.x[:,-1], simul.y[:,-1], color=domaincolor,alpha=domainalpha, linewidth=5, transform=ccrs.PlateCarree())
```

```python

plt.clf(); fig = plt.figure(figsize=(10.0,10.0)) 
projection = ccrs.Orthographic(simul.x.mean(), simul.y.mean())

ax = plt.axes(projection=projection)

ax.coastlines(resolution='110m')
ax.gridlines(draw_labels=True)

ax.stock_img(); 

####

plot_domain(simul,ax)

plt.pcolormesh(simul.x, simul.y, ma.masked_where(np.isnan(simul.mask),simul.topo),\
               cmap = plot.ncview_colormap('blu_red'),vmin=0.,vmax=8000., transform=ccrs.PlateCarree());

####

plt.savefig('gigatl3_domain_example.png', size=None, figure=None, bbox_inches='tight',\
            magnification='auto',transparent='true', dpi=300)


```
