```python
import numpy as np
import matplotlib.pyplot as plt

#################

with np.load('geostrophicadj_flat_H01_forcingallstates.npz') as data:
    times = data['arr1']
    xC    = data['arr2']
    yC    = data['arr3']
    hC    = data['arr4']
    H     = data['arr5']
    uU    = data['arr6']
    vV    = data['arr7']

g = 1

################

def avx(z):
    return (z[:, 1:]+z[:, :-1])*0.5

def avy(z):
    return (z[1:, :]+z[:-1, :])*0.5

#################

energy = 0.5*(g*(2*H*hC+hC**2)+(H+hC)*(avx(uU**2)+avy(vV**2)))

#################

plt.figure()
plt.plot(times, np.sum(energy, axis=(1, 2)))
```
